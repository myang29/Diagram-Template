```python
# Data preparation

'''
replace with own data

'''
import numpy as np
X = [10,20,30,40,50]
Y1 = list(np.random.uniform(0,1,size = 5))
Y2 = list(np.random.uniform(0,1,size = 5))
Y3 = list(np.random.uniform(0,1,size = 5))
Y4 = list(np.random.uniform(0,1,size = 5))
Y5 = list(np.random.uniform(0,1,size = 5))
Y6 = list(np.random.uniform(0,1,size = 5))
```

```python
from sklearn import datasets
from bokeh.io import output_notebook
from bokeh.plotting import figure, show, output_file


# output diagrams in notebook directly or generate html files (output_file("line chart.html"))
output_notebook()


fig = figure(title = "title",
             plot_height = 400, plot_width = 600,
             x_axis_label = "x_axis_label", y_axis_label = "y_axis_label",
             y_range = (0,1), toolbar_location = None)


# Possible line types: solid, dashed, dotted, dotdash, dashdot

# Possible scatter types: asterisk(), circle(), circle_cross(), circle_x(), cross(), diamond(), 
# diamond_cross(), inverted_triangle(), square(), square_cross(), square_x(), triangle(), x()

#Possible Colours: see https://bokeh.pydata.org/en/latest/docs/reference/colors.html


fig.line(X,Y1,legend = "Y1",line_width = 3, line_color = "lightskyblue",line_dash = "dashed")
fig.circle(X,Y1,legend = "Y1",size = 8,color = "lightskyblue")


fig.line(X,Y2,legend = "Y2",line_width = 3,line_color = "orange",line_dash = "dotted")
fig.cross(X,Y2,legend = "Y2",size = 10,color = "orange")


fig.line(X,Y3,legend = "Y3",line_width = 3,line_color = "tomato",line_dash = "dotdash")
fig.diamond_cross(X,Y3,legend = "Y3",size = 10,color = "tomato")


fig.line(X,Y4,legend = "Y4",line_width = 3,line_color = "yellowgreen",line_dash = "dashdot")
fig.asterisk(X,Y4,legend = "Y4",size = 12,color = "yellowgreen")


fig.line(X,Y5,legend = "Y5",line_width = 3,line_color = "lightpink",line_dash = "dashed" )
fig.square(X,Y5,legend = "Y5",size = 8,color = "lightpink")


# possible location for legend: top_left, top_center, top_right, center_left, center, 
# center_right, bottom_left, bottom_center, bottom_right
fig.legend.location = "bottom_left"


show(fig)
```
