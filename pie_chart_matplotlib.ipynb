```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({'name': ['a','b','c','d','e'],
                   'value':[10,20,30,25,15]})

colors = ['#C6B3DF','#B3D88E','#8EB0D8','#F2AF9A','#F9F298']

# "explode" the 2nd slice
explode = (0, 0.1, 0, 0, 0)  
fig, ax = plt.subplots(figsize=(10,6))

# title
plt.title("title", fontsize=14, loc='center', pad=10)

ax.pie(df['value'].values.tolist(), explode=explode,colors = colors,labels=df['name'].values.tolist(),autopct='%1.1f%%',
        shadow=False, startangle=90,textprops={'fontsize': 14})

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal') 
plt.legend(ncol=1 )
plt.tight_layout()
plt.savefig('pie chart.png')
plt.show()
```

```python
df = pd.DataFrame({'name': ['a','b','c','d','e'],
                   'value':[10,20,30,25,15]})

colors = ['mediumpurple','yellowgreen','lightcoral','royalblue','lightskyblue']

# "explode" the 2nd slice
explode = (0, 0.1, 0, 0, 0)  
fig, ax = plt.subplots(figsize=(10,6))

# title
plt.title("title", fontsize=14, loc='center', pad=10)

ax.pie(df['value'].values.tolist(), explode=explode,colors = colors,labels=df['name'].values.tolist(),autopct='%1.1f%%',
        shadow=False, startangle=90,textprops={'fontsize': 14})

# draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal') 
plt.legend(ncol=1 )
plt.tight_layout()
plt.savefig('donut chart.png')
plt.show()
```
