```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

data = pd.read_csv("Iris.csv")


fig, ax = plt.subplots(figsize =(9, 7)) 
a = sns.violinplot(ax = ax, x = data["Species"],  
                  y = data["SepalLengthCm"] )

# background setting
sns.set(style='darkgrid')

# title
a.axes.set_title("Title",fontsize=20,pad = 10)

# label
a.set_xlabel("X Label",fontsize=15)
a.set_ylabel("Y Label",fontsize=15)

# tick label
a.tick_params(labelsize=12)


# set False to removing top and right borders
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)


# save plot as png file
plt.savefig('violin_plot.png') # save plot as png
plt.show()
```

# grouped violin

```python
fig, ax = plt.subplots(figsize =(9, 7)) 
tips = sns.load_dataset("tips")
a = sns.violinplot(x="day", y="total_bill", hue="smoker", data=tips, palette="muted", split=True)

# background setting
sns.set(style='whitegrid')

# title
a.axes.set_title("Title",fontsize=20,pad = 10)

# label
a.set_xlabel("X Label",fontsize=15)
a.set_ylabel("Y Label",fontsize=15)

# tick label
a.tick_params(labelsize=12)


# set False to removing top and right borders
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)


# save plot as png file
plt.savefig('grouped_violin_plot.png') # save plot as png
plt.show()


```
